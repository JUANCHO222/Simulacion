# Introducción a Matplotlib

[Matplotlib](https://matplotlib.org/) es una librería que permite la creación de figuras y 
gráficos de calidad mediante el uso de Python.
* Permite la creación de gráficos de manera sencilla y eficiente.
* Permite la integración de gráficos y figuras en Jupyter Notebook.

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Mostrar los gráficos integrados dentro de Jupyter Notebook
%matplotlib inline
```

## Representación gráfica de datos 

Si a la función de trazado se le da una matriz de datos, la usara como coordenadas en el eje
vertical, utilizara el índice de cada punto de datos en el Array como la coordenada 
horizontal.

```python
plt.plot([1, 2, 5, 7, 3, 1])
plt.show()
```

Tambien se puede proporcionar dos matrices: una para el eje horizontal y otra para el eje vertical.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) #Xmin, Xmax, Ymin, Ymax
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matemática

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Tambien puede modificarse el estilo de la gráfica para que contenga más información.

```python
plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best") #La coloca en la localización más chida.
plt.show()
```

Tambien puede crearse dos gráficas que no se superpongan. Estas gráficas se organizan en un grid y se denominan subplots.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 fila, 1 columna, 1st Subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 fila, 1 columna, 2nd Subplot
plt.plot(x, y2, 'g')

plt.show()
```

Para que las gráficas no queden tan ajustadas podemos hacer la figura más grande.

```python
plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1) # 1 fila, 1 columna, 1st Subplot
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize = 14)
plt.ylabel("y1", fontsize = 14)

plt.subplot(1, 2, 2) # 1 fila, 1 columna, 2nd Subplot
plt.plot(x, y2, 'g')
plt.xlabel("x2", fontsize = 14)
plt.ylabel("y2", fontsize = 14)

plt.show()
```

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.show()
```

### Histogramas

```python
data = (1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3)
plt.subplot(211) # 2 rows, 1 Columns, 1st Subplot
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(212) # 2 rows, 1 Columns, 2nd subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel('Value')
plt.ylabel("Frequency")

plt.show()
```

### Guardar figuras

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best")
plt.savefig("Mi_grafiquita_picuda.png", transparent = True)
```
